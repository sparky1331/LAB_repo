```python
import matplotlib.pyplot as plt
import numpy as np

f = open('lab1_lin.txt')

value = [(7000 + i*6930) for i in range(100)]
alg_time = [int(f.readline()) for i in range(100)]

print(alg_time[90])
print(value[90])

value.pop(90)
alg_time.pop(90)


plt.scatter(value, alg_time)

p = np.polyfit(value, alg_time, 1)
g = np.poly1d(p)

plt.xlabel('Количество чисел в массиве')
plt.ylabel('Время поиска')

plt.plot(value, g(value))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

f = open('lab1_bin.txt')

value = [math.log(7000 + i*4000) for i in range(100)]
alg_time = [math.log(int(f.readline())) for i in range(100)]

print(alg_time)
print(value)

# value.pop(90)
# alg_time.pop(90)


plt.scatter(value[5:], alg_time[5:])

p = np.polyfit(value[5:], alg_time[5:], 1)
g = np.poly1d(p)

plt.xlabel('Количество чисел в массиве')
plt.ylabel('Время поиска')

plt.plot(value, g(value))
```
